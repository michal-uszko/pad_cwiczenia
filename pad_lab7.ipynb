### Zadanie 1

$\mathbf{H_{0}}$: Nie ma istotnej różnicy w ocenie końcowej pomiędzy uczniami i uczennicami

$\mathbf{H_{a}}$: Istnieje istotna różnica (bez wskazania kierunku) w ocenie końcowej pomiędzy uczniami i uczennicami.

Wczytanie danych

```python
grades_data = pd.read_csv('/content/drive/MyDrive/PAD/lab7/wyniki.csv')
grades_data
```

Uczennice

```python
female_grades = grades_data['ocena_koncowa'][grades_data['plec'] == 'F']
pd.DataFrame(female_grades)
```

Uczniowie

```python
male_grades = grades_data['ocena_koncowa'][grades_data['plec'] == 'M']
pd.DataFrame(male_grades)
```

Obliczenie T-statystyki dla dwóch zmiennych (uczennice i uczniowie)

```python
# Uczennice
female_std = np.std(female_grades)
female_mean = np.mean(female_grades)
female_n = female_grades.shape[0]

# Uczniowe
male_std = np.std(male_grades)
male_mean = np.mean(male_grades)
male_n = male_grades.shape[0]

# T-statystyka
t_statistic = (female_mean - male_mean) / (np.sqrt((np.power(female_std, 2) / female_n) + (np.power(male_std, 2) / male_n)))
print("t-statistic:", t_statistic)
```

Obliczenie P-value (liczba stopni swobody dla mniejszej próbki, czyli liczebność próbki uczniów minus jeden)

```python
df = male_n - 1

p = stats.t.sf(abs(t_statistic), df=df)
print("p-value:", p)
```

Według założeń poziom istotności: $\alpha=0.05$\
Obliczona wartość prawdopodobieństwa (p-value): $p≈0.027$\
\
Ponieważ $p\leq\alpha$, odrzucamy hipotezę zerową $\mathbf{H_{0}}$.\
\
Oznacza to, że **istnieje istotna różnica (bez wskazania kierunku) w ocenie końcowej pomiędzy uczniami i uczennicami.**


### Zadanie 2

```python
zyski_data = pd.read_csv('/content/drive/MyDrive/PAD/lab7/ZyskiFirmyX.csv')
zyski_data
```

```python
px.scatter(zyski_data, "Rok", "Zysk", title="Zyski w latach 1821-2020")
```

```python
model = smf.ols(formula="Zysk ~ Rok", data=zyski_data).fit()
```

```python
zyski_data["fitted"] = model.fittedvalues

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=zyski_data["Rok"], y=zyski_data["Zysk"], name="Zyski - wartości rzeczywiste", mode="markers"))
fig.add_trace(go.Scatter(
    x=zyski_data["Rok"], y=zyski_data["fitted"], name="Zyski - wartości przewidywane"))
fig.update_layout(title="Zyski w latach 1821-2020", xaxis_title="Rok",
    yaxis_title="Zyski")
fig.show()
```

```python
print(model.summary())
```

Z wykresu widać niewielki trend spadkowy, jednakże jest on znikomy, prawie niezauważalny. Oznacza to, że model regresji liniowej wyjaśnia wariancję zmiennej objaśnianej **Zysk** praktycznie tak samo dobrze jak wartość średnia.\
\
Wnioski te pokrywają się z podsumowaniem modelu, w którym **p-value** dla zmiennej **Rok** jest bardzo duże ($P=0.76$). Jest to wartość dużo większa od zazwyczaj zakładanego poziomu istotności ($α=0.05$), więc postawienie hipotezy zerowej ($\mathbf{H_{0}}$ : Zmienna niezależna **Rok** lepiej wyjaśnia wariancję zmiennej zależnej **Zysk** niż wartość średnia) skończyłoby się jej odrzuceniem.
